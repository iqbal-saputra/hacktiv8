```python
data_borobudur=pd.read_csv('https://raw.githubusercontent.com/DianCA26/Dataset/main/Pengunjung%20Candi%20Borobudur%202020-2022.csv', delimiter=";")
```

```python
data_borobudur
```

```python
import numpy as np
```

```python
# Replace data pada kolom Domestik
data_borobudur.Domestik.replace(0,np.nan, inplace=True)
```

```python
# Replace data pada kolom Mancanegara
data_borobudur.Mancanegara.replace(0,np.nan, inplace=True)
```

```python
data_borobudur
```

```python
data_borobudur.shape
```

# DROP NA

```python
data_borobudur_copy=data_borobudur.copy()
```

```python
# Menghitung total data missing dari berdasarkan kolom/tiap kolom
data_borobudur_copy.isnull().sum()
```

```python
# Menghitung total data missing dari seluruh data
data_borobudur_copy.isnull().sum().sum()
```

```python
data_borobudur_copy.shape
```

```python
data_borobudur_copy.dropna(inplace=True)
```

```python
data_borobudur_copy.shape
```

```python
data_borobudur_copy2=data_borobudur.copy()
```

```python
data_borobudur_copy2.shape
```

```python
data_borobudur_copy2
```

```python
data_borobudur_copy2.Domestik.mean()
```

```python
data_borobudur_copy2.Mancanegara.median()
```

```python
data_borobudur_copy2.fillna({'Domestik':data_borobudur_copy2.Domestik.mean(),
                             'Mancanegara':data_borobudur_copy2.Mancanegara.median()},inplace=True)
```

```python
data_borobudur_copy2
```

```python
data_borobudur_copy2.isnull().sum()
```

```python
data_borobudur_copy2.shape
```
