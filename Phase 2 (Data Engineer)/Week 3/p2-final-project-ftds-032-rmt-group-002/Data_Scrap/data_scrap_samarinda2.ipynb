```python
# Inisialisasi driver dengan link
driver = webdriver.ChromiumEdge()
url="https://www.rumah123.com/jual/rumah/?q=Samarinda"
driver.get(url)
html = driver.page_source

judul = []
hyperlink = []
img_hyperlinks = []

# Looping untuk scraping judul, hyperlink dan gambar
for i in range(1,27):
    url='https://www.rumah123.com/jual/rumah/?q=samarinda&page='+str(i)
    print(url)
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    get_id = soup.find_all('div', class_='ui-organism-intersection__element intersection-card-container')

    for container in get_id:
        a_tag = container.find('a', href=True)
        if a_tag:
            hyperlink.append("https://www.rumah123.com/"+a_tag['href'])
            judul.append(a_tag['title'])  # Menambahkan judul jika ada

        img_tag = container.find('img', src=True)
        if img_tag:
            img_hyperlinks.append(img_tag['src'])

# Menutup driver setelah selesai
driver.quit()
```
